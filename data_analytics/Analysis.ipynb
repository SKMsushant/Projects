```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns',None)
from warnings import filterwarnings
filterwarnings('ignore')
from data_ingestion_to_sql import connection_engine_to_db
```

```python
def dtype_str_category_converter(df_master):
    potential_cols=df_master.select_dtypes(exclude=['datetime']).columns.to_list()
    for col in potential_cols:
        unique_count=df_master[col].nunique()
        static_threshold=380
        if unique_count<=static_threshold:
            df_master[col]=df_master[col].astype('category')
        else:
            if (df_master[col].dtype=='object') or (df_master[col].dtype=='str'):
                df_master[col]=df_master[col].astype('object')
    return df_master
```
